Soal no. 2
Lakukan tahapan pada soal 1 dengan menggunakan tensorflow

```python
#Buat variabel skalar x dan simpan suatu nilai skalar
import tensorflow as tf
x_tf = tf.Variable(56)
x_tf
```

```python
#Tampilkan tipe data dari x
print(x_tf.dtype)
```

```python
#Buat variabel skalar y dan simpan suatu nilai skalar
y_tf = tf.Variable(-6)
y_tf
```

```python
#Tampilkan tipe data dari y
print(y_tf.dtype)
```

```python
#Jumlahkan x+y
x_tf + y_tf
```

```python
#Buat variabel bernama x_float dan simpan suatu nilai desimal
x_float_tf = tf.Variable(56, dtype=tf.float32)
x_float_tf
```

```python
#Jumlahkan x_float dengan y namun, y harus diubah tipe datanya harus disamakanya
x_float_tf + tf.cast(y_tf, dtype=tf.float32)
```

```python
#Tampilkan tipe data dari x_float
print(x_float_tf.dtype)
```
